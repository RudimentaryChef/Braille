```python
#*******************************************************************

#@author: Adi + Natalie

#Start Date: 6/28/22

#@purpose: Converts handwritten or printed text images to Braille representations, then sends it to the machine!

#*******************************************************************

#Imports
from PIL import Image # to work with images
import pytesseract # AI to scrape the text from the images
import serial # to communicate with the Arduino's serial monitor
import time # for timing

#CHANGE THESE VARIABLES DEPENDING ON YOUR ARDUINO AND COMPUTER
imga = '/Users/adi/Desktop/home.png' # Image to Translate
comPort = 'COM5' # whatever port your Arduino is on
baud = 9600 # serial monitor setting
pytesseract.pytesseract.tesseract_cmd = 'C:\\Program Files\\Tesseract-OCR\\tesseract.exe' #PATH for pytesseract

#Braille Dictionary
brailleDictionary = {"a" : "1", "b" : "12", "c" : "14",
                     "d" : "145", "e" : "15", "f" : "124",
                     "g" : "1245", "h" : "125", "i" : "24",
                     "j" : "245", "k" : "13", "l" : "123",
                     "m" : "134", "n" : "1345", "o" : "135",
                     "p" : "1234", "q" : "12345", "r" : "1235",
                     "s" : "234", "t" : "2345", "u" : "136",
                     "v" : "1236", "w" : "2456", "x" : "1346",
                     "y" : "13456", "z" : "1356",  "#" : "3456",
                     "1" : "1", "2" : "12", "3" : "14", "4" : "145",
                     "5" : "15", "6" : "124", "7" : "1245", "8" : "125",
                     "9" : "24", "0" : "245", "," : "2", ";" : "23",
                     ":" : "25", "." : "256", "?" : "236", "!" : "235",
                     " ":" ", "\n":""}

#Enhances image and scrapes the text from it
def imageToText(image):
   
    # adds more image processing capabilities
    from PIL import Image, ImageEnhance
   
    # assigning an image from the source path
    img = Image.open(imga)
   
    # adding some sharpness and contrast to the image
    enhancer1 = ImageEnhance.Sharpness(img)
    enhancer2 = ImageEnhance.Contrast(img)
    img_edit = enhancer1.enhance(20.0)
    img_edit = enhancer2.enhance(1.5)
   
    # save the new image
    img_edit.save("edited_image.png")
   
    # converts the image to result and saves it into result variable
    result = pytesseract.image_to_string(img_edit)
   
    return result.lower()

#Converts the text to a list of braille representations using the braille dictionary
def textToBraille(text):
   
    # list with all the representations
    brailleText = []
   
    # only convert symbols in our dictionary
    for i in text:
        try:
            brailleText.append(brailleDictionary[i])
        except:
            brailleText.append("")
   
    return brailleText
       

# make a list of braille representations (strings) for the words in an image
brailleList = textToBraille(imageToText(imga))

# send each of the strings to the Arduino
""" # saving this for later
for i in brailleList:
    with serial.Serial(comPort, baud, timeout=1) as ser:
        print(i.encode())
        ser.write(i.encode())    
        ## print(brailleList[i])
"""

# send the first string to the Arduino
with serial.Serial(comPort, baud, timeout=1) as ser:
    ser.write(brailleList[0].encode());
print(brailleList[0])
```
